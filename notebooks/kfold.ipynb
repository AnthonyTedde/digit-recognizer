# Analysis of kfolds

```python
# Libaries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
data = pd.read_feather("../data/train_folds.feather")

y = data.loc[:, ["label", "kfold"]]

summ = data.groupby(["kfold", "label"])["label"].count()

#print(summ.index)
nlabel = [s.values for a, s in summ.groupby(level=0)]
xlabel = [a for a, s in summ.groupby(level=0)]

lbl = list(zip(nlabel[0], nlabel[1], nlabel[2], nlabel[3], nlabel[4]))
fld0 = lbl[0]
fld1 = lbl[1]
fld2 = lbl[2]
fld3 = lbl[3]
fld4 = lbl[4]

x = np.arange(len(xlabel))
width = .19

fig, ax = plt.subplots(figsize=(15, 5))
rect1 = ax.bar(x - 2 * width, fld0, width, label="fld0")
rect2 = ax.bar(x - 1 * width, fld1, width, label="fld1")
rect3 = ax.bar(x - 0 * width, fld2, width, label="fld2")
rect4 = ax.bar(x + 1 * width, fld3, width, label="fld3")
rect5 = ax.bar(x + 2 * width, fld4, width, label="fld4")

ax.set_ylabel("Number of label")
ax.set_xlabel("fold")
ax.set_xticks(x)
ax.set_xticklabels(xlabel)
ax.legend

ax.bar_label(rect1, padding=14)
ax.bar_label(rect2, padding=3)
ax.bar_label(rect3, padding=3)
ax.bar_label(rect4, padding=3)
ax.bar_label(rect5, padding=3)

ax.legend()
#fig.tight_layout
plt.show()



```
